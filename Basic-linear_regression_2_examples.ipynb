Libraries imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

Add `x_train` and `y_train`, following ths chart
| Size (1000 sqft) | Price (1000s of dollars) |
|-----------------|--------------------------|
| 1.5             | 350                      |
| 2.5             | 550                      |

```python
x_train = np.array([1.5, 2.5])
y_train = np.array([350, 550])

print(f"x = {x_train}. y = {y_train}")
```

Use plt to create an image of the training examples.

```python
plt.scatter(x_train, y_train, c='g', marker='x')
plt.title("House prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (1000s $)")
plt.show()
```

Access  to the i'th learning example

```python
i = 0

x_i = x_train[i]
y_i = y_train[i]

print(f"x_i = {x_i}, y_i = {y_i}")
```

Set w and b values

```python
w = 200
b = 50
print(f"w = {w}, b = {b}")
```

Compute hypotesis

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

Get f_wb

```python
f_wb =  compute_model_output(x_train, w, b)
print(f_wb)

print(y_train)
```

Create plt to create image of hypotesis, training examples, title,etc

```python
# Draw hypotesis
plt.plot(x_train, f_wb,c='brown', label='Predictions')

# Scatter training examples
plt.scatter(x_train, y_train, marker='x', c='g', label="Training examples")

# Set title 
plt.title('Simple linear regression with 2 training examples')

# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')

plt.legend()
plt.show()
```
