# Basic cost function

## Goal:
Using the following training examples
| x_train | y_train |
|--------|--------|
| 1.0    | 250    |
| 1.7    | 300    |
| 2.0    | 480    |
| 2.5    | 430    |
| 3.0    | 630    |
| 3.2    | 730    |

Prompt  the user for w  and b, then, generate the hypotesis line using maatploit and print the output of the cost function

Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Load training examples

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480, 430, 630, 730])
```

Scatter the training examples

```python
plt.scatter(x_train, y_train, marker="x", color="g", label="Training examples")
plt.legend()
plt.title("House price training examples")
plt.ylabel("Price")
plt.xlabel("Size")
plt.show()
```

Get value  of m

```python
m = x_train.shape[0]
```

Gets hypotesis from all training exaples. Returns a numpy arrray.

```python
def calculate_hypotesis(x, w, b):
    return w * x + b
```

Gets the cost of the hypotesis

```python
def calculate_cost(x, y, w, b):
    f_wb = calculate_hypotesis(x, w, b)
    errors = f_wb - y

    return np.sum(errors ** 2) / (2 * m), f_wb
```

```python
def draw_hypotesis(x, y, f_wb):
    plt.scatter(x, y, color="g", marker="x", label="Training examples")
    plt.plot(x, f_wb, color="brown",label="Hypotesis")
    plt.legend()
    plt.show()
```

Asks for w and b and calculates the cost

```python
w = float(input("Enter w: "))
b = float(input("Enter b: "))
cost, f_wb = calculate_cost(x_train, y_train, w, b)
print(f"Cost is {cost}")

draw_hypotesis(x_train, y_train, f_wb)
```
